```python
# df.drop will drop the rating column and inplace will modify it permanently in data
df.drop(columns=['rating'], inplace=True)
```

```python
# df.columns IS the index of names, so call .str directly on it
df.columns = df.columns.str.capitalize()
```

```python

drop_dup = df.duplicated().sum()
drop_dup
```

```python
#df.duplicate usinng for cleaning all duplicate data fro dataset
drop_dupsdf = df.drop_duplicates()
```

```python
# finding null data for cleaning
missing_counts = df.isnull().sum()
```

```python
# drop rows with missing values
df.dropna(subset=['Date_added','Duration'], inplace=True)
```

```python
# df.fillna use for fill missing data
df.fillna({'Director' : 'Unknown', 'Cast' : 'Not Listed' , 'Country' : 'Not Available'} , inplace= True)
```

```python
# df.rename used here for renaming the 'listed_in' to 'Genre'
df.rename(columns={'Listed_in': 'Genre'}, inplace=True)
```

```python
yearly=df['Date_added'].dt.year.value_counts().sort_index()
#plt.figure(figsize=(10,8))
plt.bar(yearly.index,yearly.values)
#plt.margins(x=0.01)
plt.xticks(yearly.index,rotation=45)
plt.title("Content Added Per Year")
plt.ylabel("Count")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```
